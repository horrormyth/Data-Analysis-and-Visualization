```python
# define the childre and split the male and female
def male_female_child(passenger):
    age,sex=passenger
    if age<16:
        return 'child'
    else:
        return sex
```

```python
# make a new column wiht above defintion of child
titanic_df['Person'] = titanic_df[['Age','Sex']].apply(male_female_child,axis=1)
```

```python
# Facet Grid to plot multiple KDE Plot
fig = sns.FacetGrid(titanic_df,hue='Sex',aspect =4)
# use mapt to plot all possible kde plot by the hue choice
fig.map(sns.kdeplot,'Age',shade=True)
# set x max limit to oldest passenger
oldest =titanic_df['Age'].max()
fig.set(xlim=(0,oldest))
# Add legend
fig.add_legend()
```

```python
# 2. What deck were the passengers and how does that relate to their class

titanic_df.head()
```

```python
# Remove Nan Values from the 'Cabin'
deck = titanic_df['Cabin'].dropna()
```

```python
deck.head()
```

```python
# Extract the first letter of the deck
levels =[]
for level in deck:
    levels.append(level[0])
cabin_df = DataFrame(levels)
cabin_df.columns = ['Cabin']
sns.factorplot('Cabin',data=cabin_df,palette='winter_d',kind='count')
```

```python
# Remove inappropriate T data
cabin_df = cabin_df[cabin_df.Cabin!='T']
sns.factorplot('Cabin',data=cabin_df,palette='summer',kind='count')
```

```python
titanic_df.head()
```

```python
# 3. Where did the passengers come from
sns.factorplot('Embarked', data=titanic_df, hue='Pclass',order=['C','Q','S'],kind='count')
```

```python

# 4. Who was alone and who was with the family
# Start with adding column, defining what alone is

#sibsp column refers to sibbling, 1 being true 0 being false
#Parch is for Parents or children, 1 being true 0 being false
```

```python
titanic_df['AloneCat']= titanic_df.SibSp + titanic_df.Parch
# Results 0 being alone and 1 being with family member
```

```python
titanic_df['AloneCat']
```

```python
titanic_df.head()
```

```python
# If the summed up value is greater than  0 set it as with Family otherwise alone
titanic_df['AloneCat'].loc[titanic_df['AloneCat']>0] ='With Family'
titanic_df['AloneCat'].loc[titanic_df['AloneCat']==0] ='Alone'
```

```python
sns.factorplot('AloneCat',data=titanic_df,palette='Blues',kind='count')
```

```python
# 5. What factors helped someone survive the sinking
```

```python
# map the survivor with yes or now and set it to Survivor column
titanic_df['Suvivor'] = titanic_df.Survived.map({0:'no',1:'yes'})
```

```python
# who survived vs who died
```

```python
sns.factorplot('Suvivor',data=titanic_df,palette='Set1',kind='count')
```

```python
# what factors? check with class
sns.factorplot('Pclass','Survived',data=titanic_df)
```

```python
# Class and gender
sns.factorplot('Pclass','Survived',data=titanic_df,hue='Person')
```

```python
# Age vs Survival
```

```python
sns.lmplot('Age','Survived',data=titanic_df)
```

```python
# Effect of class on Age
sns.lmplot('Age','Survived',hue='Pclass',data=titanic_df,palette='winter')
```

```python
# Cleaning up plot with binnings
age_generations = [10,20,40,60,80]
sns.lmplot('Age','Survived',hue='Pclass',data=titanic_df,palette='winter',x_bins=age_generations)
```

```python
# gender and age with surviva
sns.lmplot('Age','Survived',hue='Sex',data=titanic_df,palette='summer',x_bins=age_generations)
```

```python
# #  1. Did the deck have an effect on the passengers survival rate ? 

# # put levels into  a new column by calling deck_Level functiond
def decklevel(cabin):
    deck = cabin
    for level in deck:
        levels =level[0]
        return levels
titanic_df['Deck']= titanic_df[['Cabin']].dropna().apply(decklevel,axis=1)
titanic_df= titanic_df[titanic_df.Cabin!='T']
sns.factorplot('Deck','Survived',data=titanic_df)
```

```python
sns.factorplot('Deck',hue='Suvivor',data=titanic_df,hue_order=['yes','no'],kind='count')
```

```python
sns.factorplot('Deck',hue='Suvivor',data=titanic_df,order=['A','B','C','D','E','F'],kind='count')
```
